```python
import pandas as pd
import math
```

```python
experiment_no=[]
for i in range (1,65):
    experiment_no.append(i)
experiment_no
```

```python
Experiment_symbol="E"
csv_files=[]
for i in range (0,64):
    csv_files.append(Experiment_symbol+str(experiment_no[i])+".csv")
csv_files
```

```python
col_names=["Time","Ax","Ay","Az"]
def Net_Accleration(x):
    return math.sqrt(x)
mean_exp_accleration=[]
```

```python
for i in csv_files:
    df=pd.read_csv(i,names=col_names)
    df["Net_Accleration"]=df["Ax"]*df["Ax"]+df["Ay"]*df["Ay"]+df["Az"]*df["Az"]
    df["Net_Accleration"]=df["Net_Accleration"].apply(Net_Accleration)
    print(df.head(4))
    mean_exp_accleration.append(df["Net_Accleration"].mean())
```

```python
mean_exp_accleration
```

```python
index_list=[]
for i in range (1,65):
    index_list.append(i)
index_list
```

```python
import pandas as pd
d={"net_accleration":mean_exp_accleration}
df_net_accleration=pd.DataFrame(d,index=index_list)
df_net_accleration
```

```python
df=pd.read_excel("Surface Roughness.xlsx")
df.set_index("Exp No", inplace=True)
df
```

```python
df=df.join(df_net_accleration)
df
```

```python
def class_acc(x):
    if x>=2.5:
        return 0
    else:
        return 1
df["class"]=df["net_accleration"].apply(class_acc)
df[22:28]
```

```python
df.rename(columns={"Average":"Roughness"},inplace=True)
df
```

```python
df2=df.dropna()
df2=df2[["Roughness","net_accleration","class"]]
df2
```

```python
from sklearn.model_selection import train_test_split

X_Train1, X_Test1, Y_Train1,Y_Test1 = train_test_split(X, Y, test_size = 0.5,random_state = 42)
X_Train2, X_Test2, Y_Train2,Y_Test2 = train_test_split(X, Y, test_size = 0.6,random_state = 42)
X_Train3, X_Test3, Y_Train3,Y_Test3 = train_test_split(X, Y, test_size = 0.3,random_state = 42)
```

```python
print("Training data: ",X_Train1.shape)
print("Testing data: ",X_Test1.shape)
print("Training data: ",X_Train2.shape)
print("Testing data: ",X_Test2.shape)
print("Training data: ",X_Train3.shape)
print("Testing data: ",X_Test3.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
X_Train1=sc_X.fit_transform(X_Train1)
X_Test1=sc_X.transform(X_Test1)
X_Train2=sc_X.fit_transform(X_Train2)
X_Test2=sc_X.transform(X_Test2)
X_Train3=sc_X.fit_transform(X_Train3)
X_Test3=sc_X.transform(X_Test3)
```

```python
from sklearn.svm import SVC
classifier11 =SVC(kernel='linear')
classifier11.fit(X_Train1,Y_Train1)
Y_Pred11=classifier11.predict(X_Test1)
classifier12 =SVC(kernel='rbf')
classifier12.fit(X_Train1,Y_Train1)
Y_Pred12=classifier12.predict(X_Test1)
classifier21 =SVC(kernel='linear')
classifier21.fit(X_Train2,Y_Train2)
Y_Pred21=classifier21.predict(X_Test2)
classifier22 =SVC(kernel='rbf')
classifier22.fit(X_Train2,Y_Train2)
Y_Pred22=classifier22.predict(X_Test2)
classifier31 =SVC(kernel='linear')
classifier31.fit(X_Train3,Y_Train3)
Y_Pred31=classifier31.predict(X_Test3)
classifier32 =SVC(kernel='rbf')
classifier32.fit(X_Train3,Y_Train3)
Y_Pred32=classifier32.predict(X_Test3)
print(Y_Pred11)
print(Y_Pred21)
print(Y_Pred31)
print(Y_Pred12)
print(Y_Pred22)
print(Y_Pred32)
```

```python
from sklearn import metrics
print('Accuracy Score:with linear kernel')
print(metrics.accuracy_score(Y_Test1,Y_Pred11))
print(metrics.accuracy_score(Y_Test2,Y_Pred21))
print(metrics.accuracy_score(Y_Test3,Y_Pred31))
print('Accuracy Score:with rbf kernel')
print(metrics.accuracy_score(Y_Test1,Y_Pred12))
print(metrics.accuracy_score(Y_Test2,Y_Pred22))
print(metrics.accuracy_score(Y_Test3,Y_Pred32))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
split=[0.5,0.6,0.3]
for i in split:
    X = df2[['net_accleration', 'Roughness']]
    y = df2['class']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=i, random_state=42)
    svm = SVC(kernel='rbf')
    svm.fit(X_train, y_train)

    y_pred = svm.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    print('Accuracy score: {:.2f}%'.format(accuracy * 100))

    plt.scatter(X_train.iloc[:, 0], X_train.iloc[:, 1], c=y_train, cmap='viridis')
    ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 100),
                     np.linspace(ylim[0], ylim[1], 100))
    Z = svm.decision_function(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)


    ax.contour(xx, yy, Z, levels=[-1, 0, 1], alpha=0.5,linestyles=['--', '-', '--'], colors='blue')
    ax.set_title('SVM with RBF kernel with train test split {}'.format(i))
    plt.show()

```
